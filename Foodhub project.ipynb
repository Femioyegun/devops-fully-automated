```python
import pandas as pd

def handle_missing_values(dataset_path):
    try:
        # Read the dataset using pandas
        df = pd.read_csv(dataset_path)

        # Check for missing values
        missing_values = df.isnull().sum()

        if missing_values.sum() == 0:
            print("No missing values found in the dataset.")
        else:
            print("Missing values:")
            print(missing_values)

            # Handling missing values
            # Option 1: Drop rows with missing values
            df_dropped = df.dropna()
            print("Dataset after dropping rows with missing values:")
            print(df_dropped)

            # Option 2: Fill missing values with mean
            df_filled_mean = df.fillna(df.mean())
            print("Dataset after filling missing values with mean:")
            print(df_filled_mean)

            # Option 3: Fill missing values with mode
            df_filled_mode = df.fillna(df.mode().iloc[0])
            print("Dataset after filling missing values with mode:")
            print(df_filled_mode)

            # Option 4: Fill missing values with forward fill
            df_filled_ffill = df.ffill()
            print("Dataset after filling missing values with forward fill:")
            print(df_filled_ffill)

            # Option 5: Fill missing values with backward fill
            df_filled_bfill = df.bfill()
            print("Dataset after filling missing values with backward fill:")
            print(df_filled_bfill)

    except FileNotFoundError:
        print("File not found.")
    except pd.errors.ParserError:
        print("Error parsing the dataset.")
    except Exception as e:
        print("An error occurred:", str(e))

# Example usage
dataset_path = "/content/drive/MyDrive/foodhub project/foodhub_order.csv"
handle_missing_values(dataset_path)
```

**No missing values found in the dataset.**

**Question 12** :
 In order to perform multivariate analysis, we need to analyze the relationships between the different variables. We can use scatter plots, correlation matrices, and groupby to achieve this.

Firstly, we handle the numerical variables. We can create a scatter plot and correlation matrix for the numerical variables to check for any linear relationships.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# assuming that df is your DataFrame
# df = pd.read_csv('your_file.csv')

# handling the rating as numerical data
df['rating'] = df['rating'].replace('Not given', None).astype(float)

# define numerical variables
numerical_vars = ['cost_of_the_order', 'food_preparation_time', 'delivery_time', 'rating']

# plot pairwise relationships in a dataset
sns.pairplot(df[numerical_vars])
plt.show()

# calculate and display correlation matrix
correlation = df[numerical_vars].corr()
sns.heatmap(correlation, annot=True)
plt.show()
```

**Below is Scatter plot and correlation matrix for numerical variables.**

* Now, for analyzing the relationships between numerical and categorical variables we can use the groupby function and visualize the data using box plots






* This will print the average cost, preparation time, and delivery time for each day of the week, and display box plots to visualize these relationships.










```python
# Grouping data by 'day_of_the_week' and calculating average cost, prep and delivery time
grouped_data = df.groupby('day_of_the_week')[numerical_vars].mean()

# print grouped data
print(grouped_data)

# Visualizing the data using box plots
for column in numerical_vars:
    plt.figure(figsize=(10, 4))
    sns.boxplot(x='day_of_the_week', y=column, data=df)
    plt.title(f'{column} by day of the week')
    plt.show()
```

**Question 13 :**

 We can use pandas' groupby function to group our data by restaurant, then we can compute the size of each group (to get the count of ratings) and the mean of 'rating' in each group. We then filter the groups by the given conditions: count > 50 and average rating > 4.

This code will print out the restaurants with more than 50 ratings and an average rating greater than 4.

```python
import pandas as pd

# assuming that df is your DataFrame
# df = pd.read_csv('your_file.csv')

# replacing 'Not given' with None and converting to float
df['rating'] = df['rating'].replace('Not given', None).astype(float)

# group by restaurant_name, compute size and mean of each group
grouped = df.groupby('restaurant_name').agg({'rating': ['size', 'mean']})

# reset column names
grouped.columns = ['rating_count', 'average_rating']

# filter restaurants with more than 50 ratings and average rating greater than 4
selected_restaurants = grouped[(grouped['rating_count'] > 50) & (grouped['average_rating'] > 4)]

print(selected_restaurants)
```

**Question 14 :**

This code will print out the total net revenue generated across all orders, taking into account the charges applied. Note that for orders costing less or equal to 5 dollars, I'm assuming the restaurant charges nothing, so the net revenue is the total cost of the order. If the charging scheme is different, you may want to adjust the lambda function accordingly.

**You can calculate the net revenue by applying the conditions stated to each row of your DataFrame and then summing up all the revenues. Here's how to do it using Python and pandas:**

```python
# assuming df is your DataFrame
# df = pd.read_csv('your_file.csv')

# calculating net revenue for each order
df['revenue'] = df['cost_of_the_order'].apply(lambda cost: cost*0.75 if cost > 20 else cost*0.85 if cost > 5 else cost)

# calculating total net revenue
total_revenue = df['revenue'].sum()

print(f"The total net revenue generated across all orders is: ${total_revenue:.2f}")
```

**The total net revenue generated across all orders is: $25148.52**

**Question 15:**

To analyze the total time required for food delivery by the restaurant, we can sum up the food_preparation_time and delivery_time for each order. Then we can calculate the percentage of orders that take more than 60 minutes to get delivered from the time the order is placed. Here's the Python code using pandas to do that:

This code calculates the total delivery time for each order and then finds out the percentage of orders that take more than 60 minutes. The result is printed out with two decimal places of precision.

```python
# assuming df is your DataFrame
# df = pd.read_csv('your_file.csv')

# calculate total time required to deliver the food
df['total_time'] = df['food_preparation_time'] + df['delivery_time']

# calculate the percentage of orders taking more than 60 minutes
over_60_minutes = df[df['total_time'] > 60].shape[0]
total_orders = df.shape[0]
percentage_over_60 = (over_60_minutes / total_orders) * 100

print(f"{percentage_over_60:.2f}% of orders take more than 60 minutes to get delivered from the time the order is placed.")
```

**10.54% of orders take more than 60 minutes to get delivered from the time the order is placed.**
