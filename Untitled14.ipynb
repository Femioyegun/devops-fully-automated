**ELBOW Method**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Sample data
data = {
    "Avg_Credit_Limit": [10000, 20000, 15000, 18000, 22000, 25000, 27000, 29000],
    "Total_Credit_Cards": [2, 4, 3, 5, 6, 7, 8, 9],
    "Total_visits_bank": [1, 2, 1, 3, 4, 2, 3, 4]
}
df = pd.DataFrame(data)

# It's usually a good idea to scale the data for k-means clustering
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

# Using the Elbow method to find a good number of clusters using WCSS
wcss = []

for i in range(1, 8):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)

# Plot the Elbow method graph
plt.figure(figsize=(10,5))
plt.plot(range(1, 8), wcss, marker='o', linestyle='--')
plt.title('K-means Clustering: The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Sample data
data = {
    "Avg_Credit_Limit": [10000, 20000, 15000, 18000, 22000, 25000, 27000, 29000],
    "Total_Credit_Cards": [2, 4, 3, 5, 6, 7, 8, 9],
    "Total_visits_bank": [1, 2, 1, 3, 4, 2, 3, 4]
}
df = pd.DataFrame(data)

# Scaling the data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

# Finding the sum of squared distances (inertia) for different k values
distortions = []
K = range(1, 8)
for k in K:
    kmeanModel = KMeans(n_clusters=k, random_state=42)
    kmeanModel.fit(df_scaled)
    distortions.append(kmeanModel.inertia_)

# Plotting the Elbow method
plt.figure(figsize=(10,5))
plt.plot(K, distortions, marker='o', linestyle='--')
plt.title('K-means Clustering: Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()

# Using an optimal k value (let's assume 3 for this example)
k_optimal = 3
kmeanModel = KMeans(n_clusters=k_optimal, random_state=42)
cluster_labels = kmeanModel.fit_predict(df_scaled)

# Add cluster labels to the original dataframe
df['Cluster'] = cluster_labels

# Display dataframe with cluster labels
print(df)

```

**SILHOUETTE SCORE Method**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Dataset
data = {
    'Avg_Credit_Limit': [26000, 9000, 68000, 9000, 10000, 18000, 183000, 19000, 10000, 13000],
    'Total_Credit_Cards': [4, 5, 6, 2, 4, 4, 9, 3, 4, 4],
    'Total_visits_bank': [5, 4, 4, 1, 5, 3, 0, 1, 2, 1],
    'Total_visits_online': [2, 1, 2, 5, 1, 2, 9, 5, 4, 5],
    'Total_calls_made': [4, 3, 2, 8, 3, 0, 2, 7, 6, 7]
}

df = pd.DataFrame(data)

# Initialize values
range_n_clusters = list(range(2, 10))  # Checking for 2 to 9 clusters
silhouette_scores = []

# Compute silhouette scores for each number of clusters
for n_clusters in range_n_clusters:
    clusterer = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = clusterer.fit_predict(df)
    silhouette_avg = silhouette_score(df, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    print(f"For n_clusters = {n_clusters}, the silhouette score is {silhouette_avg:.3f}")

# Plot the silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(range_n_clusters, silhouette_scores, 'o-')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs Number of Clusters')
plt.show()
```

**Hierarchical Clustering**

```python
import pandas as pd
from scipy.cluster.hierarchy import linkage, cophenet
from scipy.spatial.distance import pdist

# Sample dataset
data = {
    'Avg_Credit_Limit': [26000, 9000, 68000, 9000, 10000, 18000, 183000, 19000, 10000, 13000],
    'Total_Credit_Cards': [4, 5, 6, 2, 4, 4, 9, 3, 4, 4],
    'Total_visits_bank': [5, 4, 4, 1, 5, 3, 0, 1, 2, 1],
    'Total_visits_online': [2, 1, 2, 5, 1, 2, 9, 5, 4, 5],
    'Total_calls_made': [4, 3, 2, 8, 3, 0, 2, 7, 6, 7]
}

df = pd.DataFrame(data)

# Different distance metrics and linkage methods
distance_metrics = ['euclidean', 'cityblock', 'cosine']
linkage_methods = ['single', 'complete', 'average', 'ward']

for dist_metric in distance_metrics:
    for link_method in linkage_methods:
        if link_method == 'ward' and dist_metric != 'euclidean':
            # The 'ward' linkage method only works with the 'euclidean' distance metric
            continue
        Z = linkage(df, metric=dist_metric, method=link_method)
        c, _ = cophenet(Z, pdist(df))
        print(f"Cophenetic correlation for {dist_metric} distance with {link_method} linkage: {c:.3f}")

```

**To print the combination of distance metric and linkage method with the highest cophenetic correlation**

```python
from scipy.cluster.hierarchy import linkage, cophenet
from scipy.spatial.distance import pdist

# Define the distance metrics and linkage methods
distance_metrics = ['euclidean', 'cityblock', 'cosine']
linkage_methods = ['single', 'complete', 'average', 'ward']

# Store results in a dictionary
results = {}

# Calculate cophenetic correlation for each combination
for dist_metric in distance_metrics:
    for link_method in linkage_methods:
        if link_method == 'ward' and dist_metric != 'euclidean':
            continue  # Ward linkage can only work with euclidean distance

        linkage_matrix = linkage(df, method=link_method, metric=dist_metric)
        coph_corr, _ = cophenet(linkage_matrix, pdist(df, metric=dist_metric))
        results[(dist_metric, link_method)] = coph_corr

# Get the combination with the highest cophenetic correlation
best_combination = max(results, key=results.get)

print(f"The combination with the highest cophenetic correlation is:")
print(f"Distance Metric: {best_combination[0]}")
print(f"Linkage Method: {best_combination[1]}")
print(f"Cophenetic Correlation: {results[best_combination]}")
```

The combination with the highest cophenetic correlation is:
Distance Metric: cityblock
Linkage Method: average

**visualization using boxplot of scaled numerical variables for each clusters.**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import linkage, fcluster

# Sample dataset
data = {
    'Avg_Credit_Limit': [26000, 9000, 68000, 9000, 10000, 18000, 183000, 19000, 10000, 13000],
    'Total_Credit_Cards': [4, 5, 6, 2, 4, 4, 9, 3, 4, 4],
    'Total_visits_bank': [5, 4, 4, 1, 5, 3, 0, 1, 2, 1],
    'Total_visits_online': [2, 1, 2, 5, 1, 2, 9, 5, 4, 5],
    'Total_calls_made': [4, 3, 2, 8, 3, 0, 2, 7, 6, 7]
}

# Convert dictionary to DataFrame
data_df = pd.DataFrame(data)

# Scale the data
scaler = StandardScaler()
scaled_data = pd.DataFrame(scaler.fit_transform(data_df), columns=data_df.columns)

# Perform hierarchical clustering
Z = linkage(scaled_data, 'ward')
labels = fcluster(Z, t=2, criterion='maxclust')
scaled_data['cluster_labels'] = labels

# Visualization using box plots
plt.figure(figsize=(15, 10))

for i, column in enumerate(scaled_data.columns[:-1], 1):
    plt.subplot(2, 3, i)
    scaled_data.boxplot(column=column, by='cluster_labels')
    plt.title(column)
    plt.suptitle('')  # This line removes the default 'Boxplot grouped by cluster_labels' title

plt.tight_layout()
plt.show()
```
